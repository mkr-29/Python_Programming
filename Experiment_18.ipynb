3. For two people born on different days, there is a day when one is twice as old as the other. That’s their Double Day. Write a program that takes two birthdays and computes their Double Day. 

```python
import datetime

def double_day(b_day1, b_day2):
    if(b_day1>b_day2):
        start_date=b_day1
    else:
        start_date=b_day2

    for i in range(100*365):
        c_date=start_date+datetime.timedelta(days=i)
        d1=c_date-b_day1
        d2=c_date-b_day2
        if(d1>d2):
            if(d1.days==2*d2.days):
                print(c_date.strftime("%d-%m-%Y"))
        else:
            if(d2.days==2*d1.days):
                print(c_date.strftime("%d-%m-%Y"))
                
b_day1=input("Enter first birthday (DD-MM-YYYY)? ")
b_day1=datetime.datetime.strptime(b_day1,"%d/%m/%Y").date()  
b_day2=input("Enter second birthday (DD-MM-YYYY)? ")
b_day2=datetime.datetime.strptime(b_day2,"%d/%m/%Y").date()
double_day(b_day1, b_day2)
```

4. For a little more challenge, write the more general version that computes the day when one person is n times older than the other.

```python
import datetime

def find_n_times_day(bday1, bday2, n):
    person1 = datetime.date(*bday1)
    person2 = datetime.date(*bday2)
    age_diff = -(person1 - person2)
    p1 = int(age_diff.days)
    p2 = 0
    while p2 * n != p1:
        if p2 * n > p1:
            print("This never precisely occurred")
            return None
        p1 += 1
        p2 += 1
    date_at_n_times_age = person2 + datetime.timedelta(days=p2)
    print(date_at_n_times_age,"\n","person 1 was %d days old, and person 2 was %d days old" % (p1, p2))

tup_1=(2002,8,9)
tup_2=(2005,3,13)
find_n_times_day(tup_1,tup_2,2)
```
